### Normalización

```python
from sklearn import preprocessing
import pandas as pd
import numpy as np
normalizer = preprocessing.MinMaxScaler()
columnas=dataSetClientesSinCols.columns
dataSetClientesNorm = dataSetClientesSinCols.copy()

#Normalizar columnas no categóricas
categoricas=['card4','card6','ProductCD','id_12','id_15','id_16','id_28','id_29','id_31','id_35','id_36','id_37','id_38','DeviceType','isFraud']
continuas=columnas[np.logical_not(columnas.isin(categoricas))]
if normalizacion == True:
    for c in continuas:
        dataSetClientesNorm[c] = normalizer.fit_transform(pd.DataFrame(dataSetClientesSinCols[c]))
    dataSetClientesNorm
else:
    dataSetClientesNorm=dataSetClientesSinCols.copy()
```

```python
filePickle = open('filePickleEstruct.pkl','wb')
pickle.dump(dataSetClientesNorm,filePickle)
filePickle.close()
```

### Validación de resultados

```python
from sklearn.metrics import roc_curve, auc
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
def validarParametros(df):    
    
    metricasClas = {
      'AUC':   metrics.roc_auc_score,  
      'ACC':    metrics.accuracy_score,
    }
    
    dataX=df.drop('isFraud',axis=1)
    tarjetY=df[['isFraud']]

    X_train, X_test, y_train, y_test \
    = train_test_split(dataX, tarjetY, stratify=tarjetY, test_size=0.2, random_state=0)
    parametrosDT={'criterion': 'gini', 'max_depth': 20,'random_state':42}
    model=DecisionTreeClassifier(**parametrosDT)
    
    model.fit(X_train,y_train)
    y_pred =  model.predict(X_test)
   
        
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
    roc_auc = auc(false_positive_rate, true_positive_rate) 
    
    ACC = metricasClas['ACC'](y_test, y_pred)
   
    print("%s\n AUC: %.3f ACC: %.3f " % ('MÉTRICAS -->',roc_auc, ACC))  
```

```python
validarParametros(dataSetClientesNorm)
```
